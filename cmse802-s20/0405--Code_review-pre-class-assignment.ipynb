# 2. Code Review Checklist

&#9989; **<font color=red>DO THIS:</font>** Read the following blog post about code reviews https://www.codeproject.com/Articles/1156196/Code-Review-Checklist and come to class ready to discuss what you learned.

<font size=4 color=blue>basic code review
- Objective based [Purposeful] 
- Unbreakable [Validated] 
- Responses are handled [Error handling and Data formatting] 
- Follows architecture [No design deviation] 
- Unit tested [Reliable] 
- Reusable [No repetition] 
- Performance oriented [Speedy response & Scalability] 
- Secure [Safe code] 
- Manageable [Crisp and Formatted] 
- Meets coding conventions and standards [Standardized approach] 
- Detach connections after usage [Memory handling] 
- Has configurable logging[Traceability] 
- Code coverage is more than 95% [Efficiency] 
- On-demand resource delivery[Fast] 
- No warnings / console logs [Data safety] 
- Legal usage of third-party tools/libraries[Licencing]
